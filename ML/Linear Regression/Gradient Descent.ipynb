```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### gradient-descent implementation

```python
def grad_desc(x,y,it):
    m_curr=c_curr=0
    learning_rate=0.08
    n=len(x)
    plt.plot(x,y,color='r',linewidth=5)
    for i in range(it):
        y_pred=m_curr*x+c_curr
        plt.plot(x,y_pred,color='greenyellow')
        cost=(1/n)*sum([j**2 for j in (y-y_pred)])
        md=-(2/n)*sum(x*(y-y_pred))
        cd=-(2/n)*sum(y-y_pred)
        m_curr=m_curr-(learning_rate*md)
        c_curr=c_curr-(learning_rate*cd)
#         print(f"m: {m_curr} c: {c_curr} cost: {cost} iterations: {it}")
    print(m_curr,c_curr)
```

'''to figure out the best learning rate 
we can start with say a learning rate of 0.001 or 0.0001 and 10 iterations. 
If the cost is decreases, we can proceed further with increasing the learning rate till our cost decreases and 
after figuring out the best learning rate we can increase the iterations to lets say: 1000 -10000''' 

```python
x=np.array([1,2,3,4,5])
y=np.array([5, 7, 9, 11, 13])
iterations=10000
```

```python
grad_desc(x,y,iterations)
```

### using grad-desc on test_scores dataset

```python
import math
import pandas as pd
```

```python
def grad_desc(x,y,learning_rate,it):
    m_curr=c_curr=0
    n=len(x)
    plt.scatter(x,y,color='r',linewidth=5)
    cost_prev=0
    for i in range(it):
        y_pred=m_curr*x+c_curr
        plt.plot(x,y_pred,color='green')
        cost=(1/n)*sum([j**2 for j in (y-y_pred)])
        md=-(2/n)*sum(x*(y-y_pred))
        cd=-(2/n)*sum(y-y_pred)
        m_curr=m_curr-(learning_rate*md)
        c_curr=c_curr-(learning_rate*cd)
        if math.isclose(cost,cost_prev,rel_tol=1e-20):
            break
        cost_prev=cost
#         print(f"m: {m_curr} c: {c_curr} cost: {cost} iterations: {i}")
#     plt.scatter(x,y_pred,color='b',linewidth=5)
    return m_curr,c_curr
```

```python
from sklearn import linear_model
```

```python
def usingsk(x_sk,y_sk):
    alg=linear_model.LinearRegression()
    alg.fit(x_sk,y_sk)
    return alg.coef_,alg.intercept_
```

```python
if __name__=="__main__":
    df=pd.read_csv('test_scores.csv')
    x=df.math
    y=df.cs
    learning_rate=0.0002
    
    m,c=grad_desc(x,y,learning_rate,100000)
    print(f"m: {m} c: {c}")
    
    m_sk,c_sk=usingsk(df[['math']],y)
    print(f"m_sk: {m_sk} c_sk: {c_sk}")
```
